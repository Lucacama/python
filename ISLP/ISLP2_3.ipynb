# Matplotlib guide


```python

fig, ax = subplots(figsize=(5,5)) #figsize = physical size of graph
x = rng.standard_normal(100) 
y = rng.standard_normal(100)
ax.plot(x, y,) #ploted the random datasets of x and y
```

```python
fig, ax = subplots(figsize=(5,5))
ax.scatter(x,y); #ax.plot(x,y,'o') alternative / the semicolon stops text form printing
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
ax.set_title("x vs y")
```

```python
fig, ax = subplots(ncols=2, sharex=True, figsize = (10,5))
ax[0].scatter(x, y,)
ax[1].plot(x,y,)
```

```python
fig, ax = subplots(figsize=(5,5))
x = np.linspace(-np.pi, np.pi, 50)
y = x
z = np.multiply.outer(np.cos(y), 1 / (1 + x**2))
ax.contour(x, y, z, levels=10); # Uses colour to represent Z, use levels to change resoulution
```

```python
fig, ax = subplots(figsize=(5,5))
ax.imshow(z);
```

```python
rng = np.random.default_rng(1)
a = rng.standard_normal((127,5))
m = rng.choice([0,np.nan], p=[0.8, 0.2], size=a.shape)
a += m
d = pd.DataFrame(a, columns=['food', 'bar', 'pickle', 'snack', 'popcorn'])
d
```

```python
for col in d.columns:
    missing = np.isnan(d[col]).mean() * 100
    template = f"coulumn {col} has {missing: .2f}% missing values"  # My attempt
    print(template)
```

```python
for cols in d.columns:
    template = 'column "{0}" has {1:.2%} missing values'
    print(template.format(cols, np.isnan(d[cols]).mean())) # Better way proposed by the book, the % assigner is unique to format function
```
